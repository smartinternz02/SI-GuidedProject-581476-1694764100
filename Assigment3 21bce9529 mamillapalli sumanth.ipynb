```python
lenmedian=df['culmen_length_mm'].median();
df['culmen_length_mm'].fillna(lenmedian,inplace=True)
```

```python
depthmedian=df['culmen_depth_mm'].median()
df['culmen_depth_mm'].fillna(depthmedian,inplace=True)
```

```python
flipper=df['flipper_length_mm'].median()
df['flipper_length_mm'].fillna(flipper,inplace=True)
```

```python
bodymass=df['body_mass_g'].median()
df['body_mass_g'].fillna(bodymass,inplace=True)
```

```python
modesex=df['sex'].mode()[0]
df['sex'].fillna(modesex,inplace=True)
```

```python
df['sex']=df['sex'].replace(".",modesex)

df.sex.value_counts()
```
